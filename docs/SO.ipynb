```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carregar os dados do arquivo CSV
df = pd.read_csv('dados.csv')

# Substituir 'NULL' por NaN
df.replace('NULL', np.nan, inplace=True)

# Converter as colunas relevantes para números
df[' Tempo (Sequencial)'] = pd.to_numeric(df[' Tempo (Sequencial)'], errors='coerce')

# Remover linhas com valores NaN em ' Tempo (Sequencial)'
df.dropna(subset=[' Tempo (Sequencial)'], inplace=True)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df['Tamanho da Matriz'], df[' Tempo (Sequencial)'], marker='o', color='b', linestyle='-')
plt.title('Tempo de Sequencial em Função do Tamanho da Matriz')
plt.xlabel('Tamanho da Matriz')
plt.ylabel('Tempo (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Carregar os dados do arquivo CSV
df2 = pd.read_csv('dados.csv')
print(df2.columns)

# Substituir 'NULL' por NaN
df2.replace('NULL', np.nan, inplace=True)

# Filtrar as linhas onde 'Tempo (Processos)' não é NaN
df2 = df2[df2[' Tempo (Processos)'].notnull()]

# Converter o tipo das colunas para float
df2['P (Processos)'] = df2[' P (Processos)'].astype(float)
df2[' Tempo (Processos)'] = df2[' Tempo (Processos)'].astype(float)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df2[' P (Processos)'], df2[' Tempo (Processos)'], marker='o', color='r', linestyle='-')
plt.title('Tempo em Função de P para Processos')
plt.xlabel(' P (Processos)')
plt.ylabel('Tempo (s)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Carregar os dados do arquivo CSV
df3 = pd.read_csv('dados.csv')

# Filtrar as linhas onde 'Tempo (Threads)' não é NULL
df3 = df3[df3[' Tempo (Threads)'].notnull()]

# Converter o tipo das colunas para float
df3[' P (Threads)'] = df3[' P (Threads)'].astype(float)
df3[' Tempo (Threads)'] = df3[' Tempo (Threads)'].astype(float)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df3[' P (Threads)'], df3[' Tempo (Threads)'], marker='o', color='b', linestyle='-')
plt.title('Tempo em Função de P para Threads')
plt.xlabel(' P (Threads)')
plt.ylabel(' Tempo (s)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
